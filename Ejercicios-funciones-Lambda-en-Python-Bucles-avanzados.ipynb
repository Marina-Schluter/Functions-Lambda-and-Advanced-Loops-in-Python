# Ejercicio 7
Cree una función para contar la cantidad de números negativos de una lista.


````python
#Solución con lambda

programaneg=lambda lista: len(list(filter(lambda x:x<0,lista)))
programaneg(lista)
````

````python
#Solución con contador

#Lista de ejemplo
lista=[7,23,-45,89,-46,-32]

#Crear función para contar los negativos

def contar_negativos(lista):
    #Iniciamos contador a 0
    contador=0
    #Revisamos los números de la lista con for y le ponemos las condiciones
    for numero in lista:
        if numero<0:
            contador+=1
    return contador

#Aplicar función al ejemplo
contaje_de_negativos=contar_negativos(lista)

print(f"En la lista de números del ejemplo hay",contaje_de_negativos,"números negativos")
````

````python
#Otra solución con contador

#Lista de ejemplo
lista=[7,23,-45,89,-46,-32]

#Función para contar negativos

def contar_negativos(lista_numeros):
    if not lista_numeros:return[]
    return[len([numero for numero in lista_numeros if numero <0])]
print(contar_negativos(lista))
````

# Ejercicio 8

Cree una función para contar la cantidad de números impares de una lista.


````python
#Lista de ejemplo
lista=[7,23,-45,89,-46,-32]

#Crear función para contar los negativos

def contar_impares(lista):
    #Iniciamos contador a 0
    contador=0
    #Revisamos los números de la lista con for y le ponemos las condiciones
    for numero in lista:
        if numero%2!=0:
            contador+=1
    return contador

#Aplicar función al ejemplo
contaje_de_impares=contar_impares(lista)

print(f"En la lista de números del ejemplo hay",contaje_de_impares,"números impares")
````

# Ejercicio 9

Dada una palabra, cuente el número de vocales que tiene.



````python
palabra="australopithecus"

def contar_vocales(palabra):
    #Definimos vocales
    vocal=['a', 'e', 'i', 'o', 'u', 'á', 'é', 'í', 'ó', 'ú', 'à', 'è', 'ì', 'ò', 'ù', 'ä', 'ë', 'ï', 'ö', 'ü', 'ÿ', 'â', 'ê', 'î', 'ô', 'û']
    #Convertimos a minúsculas
    palabra=palabra.lower()
    #Iniciamos contador
    contador=0
    #Revisamos la palabra en busca de vocales
    for elemento in palabra:
        if elemento in vocal: 
            contador+=1
    return contador

contaje_vocales=contar_vocales(palabra)
print(f"La palabra",palabra,"tiene",contaje_vocales,"vocales")
    

````

# Ejercicio 10
Imprima el abecedario al revés.

Puede ayudarse de [reversed.](https://docs.python.org/3/library/functions.html?highlight=reverse#reversed)

````python
#Importamos biblioteca string
import string 

#Invocamos el alfabeto (hay 3 tipos, en mayúsculas, en minúsculas y la suma de ambas)
alfabeto_mayusculas=string.ascii_uppercase
alfabeto_minusculas=string.ascii_lowercase
alfabeto=string.ascii_letters
print(alfabeto_mayusculas)
print(alfabeto_minusculas)
print(alfabeto)

#Trabajamos con el de minúsculas y lo separamos por un guión
abecedario_separado="-".join(alfabeto_minusculas)
print(abecedario_separado)

#Usamos reversed para onterne el abecedario al revés
abecedario_al_reves=reversed(abecedario_separado)
print(list(abecedario_al_reves))
````

````python
#Opción optimizada

#Importamos biblioteca
import string

#Invocamos el abecedario desde la bblioteca
abecedario=string.ascii_lowercase
print(abecedario)

#Usamos join para obtener los datos separados por in guión en este caso y le damos la vuelta con reversed
abecedario_reves="-".join(reversed(abecedario))
print(abecedario_reves)
````

# Ejercicio de ampliación 2

Dada una matrix 

1 2 

3 4

5 6

7 8 



```
matrix = [[1, 2], [3, 4], [5, 6], [7, 8]]
```

Se pide (de forma independiente por cada apartado):

**1**) Pida un número por consola al usuario usando el siguiente comando:


```
numero = input("Por favor ingrese un número: ")
```
**2**) Controle que el número esté entre 1 y 9. Muestre un mensaje de error en caso de no complir dicha condición. El programa no debe continuar (no use `break` para realizar esta cuestión).

**3**) Encuentre aquellos factores superiores a dicho número. Para localizar un elemento de la matrix puede usar:

```
elemento = matrix[fila][columna]
```
*Pista*: Use bucles aninados, `range` y `len`.

**4**) Devuelva su matrix *traspuesta*.
[Matriz traspuesta](https://upload.wikimedia.org/wikipedia/commons/e/e4/Matrix_transpose.gif). Este punto puede realizarlo fuera de las condiciones realizadas anteriormente.

**Extra**) puede controlar si un usuario introduce otro valor (letra, símbolo u otro caracter) usando `isdigit`, `is_numeric` o `is_decimal`. ¿Cuál vendría mejor? 


**En todo momento trate de usar listas de comprensión, siempre que le sea posible.**

````python
#Matriz inicial
matrix = [[1, 2], [3, 4], [5, 6], [7, 8]]
````

````python
#Pedir al usuario que ingrese un número y mostrarlo
numero =input("Por favor ingrese un número: ")
print("Por favor ingrese un número: ",numero)

#Verificar si el número es un número y si está entre el 1 y el 9
if not numero.isdigit() or int(numero) < 1 or int(numero) > 9:
    print("Error: El número debe estar entre 1 y 9.")
else:
    #Convertir el número del usuario en número entero para poder compararlo
    numero = int(numero)
    
    #Buscar los factores superiores al número dado por el usuario y mostrarlos
    factores_superiores = [elemento for fila in matrix for elemento in fila if elemento > numero]
    print(factores_superiores)

    #Calcular la matriz traspuesta
    matriz_traspuesta = [[fila[i] for fila in matrix] for i in range(len(matrix[0]))]

    # Imprimir la matriz traspuesta
    print("Matrix traspuesta")
    for fila in matriz_traspuesta:
        print(fila)
````
